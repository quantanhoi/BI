```python
import pandas as pd
from IPython.display import display

# Read CSV with correct delimiter and decimal settings
# - sep=';' because the file uses semicolons
# - decimal=',' to parse numbers like 10669,84
# - on_bad_lines='skip' to ignore any malformed rows
# - engine='python' for more robust parsing

df = pd.read_csv(
    'SalesData.csv',
    sep=';',
    decimal=',',
    on_bad_lines='skip',
    engine='python'
)

# Parse dates (day-first format like 01.01.07)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True, errors='coerce')

# Quick preview and schema
display(df.head())
df.info()
```

# Unique pairs: Product × Division

- Unique combinations only
- Pairs with frequency (descending)
- Number of distinct pairs

```python
# Unique Product–Division pairs
import pandas as pd
from IPython.display import display

# 1) Unique combinations only
unique_pairs = df[['CustDescr', 'City', 'SalesOrg']].drop_duplicates()
print(f"Distinct pairs: {len(unique_pairs):,}")
display(unique_pairs.head(20))

# 2) Frequency of each pair (descending)
pair_counts = (
    df.value_counts(['CustDescr', 'City', 'SalesOrg'])
      .reset_index(name='count')
      .sort_values('count', ascending=False)
)
display(pair_counts.head(20))

# 3) Optional checks (uncomment as needed)
# How many divisions per product?
# divisions_per_product = df.groupby('Product')['Division'].nunique().sort_values(ascending=False)
# display(divisions_per_product.head(20))

# Products mapped to more than one Division (potential data issue)
# multi_div_products = divisions_per_product[divisions_per_product > 1]
# display(multi_div_products)
```
